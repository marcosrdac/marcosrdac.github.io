```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
```

```python
# generating globs
N = 100
c = np.asarray(
  [[1, 1],   # 1st glob center coordinates
   [4, 3],]  # 2nd glob center coordinates
)
sigma = [1.0, 0.5]
x1 = np.stack([np.random.normal(i, sigma[0], N, ) for i in c[0]], axis=1)
x2 = np.stack([np.random.normal(i, sigma[1], N, ) for i in c[1]], axis=1)
```

```python
fig, ax = plt.subplots()
ax.set_title('Normal distributions and their centroids')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.scatter(x1[:, 0], x1[:, 1], c=None)
ax.scatter(x2[:, 0], x2[:, 1], c=None)
ax.scatter(c[0, 0], c[0, 1], marker='X', s=100, c='black', label='centroids')
ax.scatter(c[1, 0], c[1, 1], marker='X', s=100, c='black')
ax.legend()
plt.savefig("distributions.svg")
```
